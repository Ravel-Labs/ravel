```python
def butter_bandpass(lowcut, highcut, fs, order):
    nyq = 0.5 * fs
    low = lowcut / nyq
    high = highcut / nyq
    b, a = butter(order, [low, high], btype='band')
    return b, a

def rms_normalization(x, R):
    """Calculate scalar for signal on a linear scale"""
    n = x.shape[0]
    a = np.sqrt((n * R**2) / np.sum(x**2))
    return a

def min_max_normalization(x, min_value, max_value):
    min_x = x.min()
    max_x = x.max()
    x_prime = (x - min_x)
    x_prime = x_prime / max_x
    x_prime = x_prime * (max_value - min_value) + min_value
    return x_prime

def mean_normalization(x):
    mu = x.mean()
    min_x = x.min()
    max_x = x.max()
    x_prime = (x - mu) / (max_x - min_x)
    return x_prime

def calc_X_w(x, n_fft, window_size, hop_length, R, a_0, a_1, fs, C, rows, n, seconds):
    x_sig = x.signal[fs*seconds*n:fs*seconds*(n+1)]
    A = rms_normalization(x_sig, R)
    x_sig = A * x_sig
    X = np.abs(librosa.stft(x_sig, n_fft=n_fft, win_length=window_size, hop_length=hop_length))
    M = X.shape[1]
    W = np.zeros((C, M))
    for c in range(52):
        f_c = 229 * (10**((a_1*c + a_0)/24) - 1)
        b_c = 0.108*f_c + 24.7
        b_half = b_c / 2
        low = f_c - b_half
        high = f_c + b_half
        b, a = butter_bandpass(low, high, fs, order=2)
        y = lfilter(b, a, x_sig)
        y_fft = np.abs(np.fft.fft(y, n=M))
        W[c] = y_fft
    X_w = np.matmul(W, X.T)
    num_idx = X_w.shape[0] * X_w.shape[1]
    if num_idx % rows == 0:
        cols = int(num_idx / 5)
    else:
        cols = int(num_idx / 5) + 1
    X_w.resize((rows, cols))
    return X_w, W

def compute_yhat(Y_w, X_w, W, R, hop_length, window_size):
    Y = np.matmul(W.T, (Y_w*R))
    X = np.matmul(W.T, (X_w*R))
    G = sigmoid(b, (Y/X))
    X_G = X * G
    Y_hat = librosa.core.istft(X_G, hop_length=hop_length, win_length=window_size)
    return Y_hat
    
    
def sigmoid(b, x): return 2 / (1 + np.exp(-b*x))

def compute_R(Y_w, X_w): return np.log10(Y_w) * (np.log10(X_w))**-1
```

```python
class Autoencoder_Xw(nn.Module):
    def __init__(self):
        super(Autoencoder_Xw, self).__init__()
        self.encoder = nn.Sequential(
        nn.Linear(5*21310, 260),
        nn.Dropout(0.3),
        nn.ReLU(True),
        nn.Linear(260, 260),
        nn.Dropout(0.3),
        nn.ReLU(True))
        self.decoder = nn.Sequential(
        nn.Linear(260, 260),
        nn.Dropout(0.3),
        nn.ReLU(True),
        nn.Linear(260, 260),
        nn.Dropout(0.3),
        nn.ReLU(True),
        nn.Linear(260, 5*21310))
        
    def forward(self, x):
        x = self.encoder(x)
        x = self.decoder(x)
        return x
    
class Autoencoder_R(nn.Module):
    def __init__(self):
        super(Autoencoder_R, self).__init__()
        self.encoder = nn.Sequential(
        nn.Linear(5*21310, 260),
        nn.Dropout(0.3),
        nn.ReLU(True),
        nn.Linear(260, 350),
        nn.Dropout(0.3),
        nn.ReLU(True))
        self.decoder = nn.Sequential(
        nn.Linear(350, 260),
        nn.Dropout(0.3),
        nn.Linear(260, 5*21310))
        
    def forward(self, x):
        x = self.encoder(x)
        x = self.decoder(x)        

class DNN(nn.Module):
    def __init__(self):
        super(DNN, self).__init__()
        self.layers = nn.Sequential(
            nn.Linear(5*21310, 260),
            nn.ReLU(True),
            nn.Linear(260, 260),
            nn.ReLU(True),
            nn.Linear(260, 260),
            nn.ReLU(True),
            nn.Linear(260, 260),
            nn.ReLU(True),
            nn.Linear(260, 5*21310),
            nn.Linear(5*21310, 260),
            nn.ReLU(True),
            nn.Linear(260, 350),
            nn.ReLU(True),
            nn.Linear(350, 260),
            nn.Linear(260, 5*21310))
        
        def forward(self, x):
            x = self.forward(x)
                    
def load_data(path, warp_path, w_path, np_save, n_fft=4096, window_size=2049, hop_length=1025, R=0.5, a_0=1.5, a_1=0.79, C=52, rows=5, n=0, seconds=30):        
    if np_save:
        files = full_file_scraper(path)
        signals = [Signal(path, n_fft=n_fft, window_size=window_size, hop_length=hop_length) for path in files]
        X_w_list = []
        W_list = []
        for sig in signals:
            fs = sig.sr
            num_segments = int(sig.signal.shape[0] / (fs * 30))
            for n in range(num_segments):
                X_w, W = calc_X_w(sig, n_fft, window_size, hop_length, R, a_0, a_1, fs, C, rows, n, seconds)
#                 X_w = mean_normalization(X_w)
                X_w = librosa.util.normalize(X_w)
                W_list.append(W)
                X_w_list.append(X_w)
        np.save(warp_path, X_w_list)
#         np.save(w_path, W_list)
    else:
        X_w_list = np.load(warp_path)
#         W_list = np.load(w_path)
    X_ws = torch.Tensor(X_w_list)
#     Ws = torch.Tensor(W_list) 
#     return X_ws, Ws
    return X_ws
                
def save_model(model, path): torch.save(model.state_dict(), path)
    
def upload_Xw_R_params(model_Xw, model_R, model_Xw_R, path_Xw, path_R):
    model_Xw = Autoencoder_Xw()
    model_Xw.load_state_dict(torch.load(path_Xw), strict=False)
    model_R = Autoencoder_R()
    model_R.load_state_dict(torch.load(path_R), strict=False)
    combined = model_Xw.state_dict()
    comined.update(model_R.state_dict())
    model_Xw_R.load_state_dict(combined)
    
def train(model_type, dataset, batch_size, num_epochs, lr, weight_decay, path, dnn, param_list=[]):
    dataloader = DataLoader(dataset, batch_size=batch_size, shuffle=True)
    model = model_type()
    criterion = nn.MSELoss()
    optimizer = torch.optim.Adam(
    model.parameters(), lr=lr, weight_decay=weight_decay)
    if dnn == True:
        model_Xw, model_R = param_list[0], param_list[1] 
        path_Xw, path_R = param_list[2], param_list[3]
        upload_Xw_R_params(model_Xw, model_R, model, path_Xw, path_R)       
    for epoch in range(num_epochs):
        for data in dataloader:
            sig = data
            sig = sig.view(sig.size(0), -1)
            sig = Variable(sig)
            output = model(sig)
            loss = criterion(output, sig)
            optimizer.zero_grad()
            loss.backward()
            torch.nn.utils.clip_grad_norm_(model.parameters(), 0.5)
            optimizer.step()
            
        print('epoch [{}/{}], loss:{:.4f}'.format(epoch + 1, num_epochs, loss.data))
    torch.save(model.state_dict(), path)
            
```
